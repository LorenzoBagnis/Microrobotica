## Intelligenza Artificiale
##### Ada Lovelace ( 1815-1852)
    figlia di un poeta inglese (Lord Byron)    
    famosa perchè ha conosciuto Charles Babbage (inventore primo computer: "macchina analitica")    
    Ada ha studiato matematica e programma questo computer su carta    
    gli è stato dedicato un linguaggio di programmazione per uso aereospaziale    
    si è posta la domanda se una macchina fosse capace di creare (intelligenza artificiale) pensava principalmente alla musica    
    citata più volte da alan turing

    legge principio della matematica di Rassel (legge in biblioteca per 3 giorni) mai laureato    
    cervello animali composto da neuroni (cellula con dendriti che lo collegano con altri neuroni, attraverso segnali elettrici

##### Alan Turing (1912-1954)
    si è suicidato mangiando una mela avvelenata (aneddoto mela apple collegata a questo suicidio)
    Premio Turing (è il premio nobel ma per gli informatici)
    una macchina è in grado di pensare?
    test di Turing verifica se una macchina è intelligente o no
    uomo chatta con una macchina il test è passato se non si accorge di stare chattando con una macchina ma pensa di parlare con un umano (test ipotetico non si ha con certezza un risultato)
    al giorno d'oggi spesso il test è svolto al contrario per verificare che si è umani (CHAPTA) serve per annullare il web scraping (bot che viaggiano su internet per rubare informazioni)
### nase in una conferenza nel 1956 (jonh McCarthy, Marvin Lee Minsky, Claude Elvood Shannon, Rochester)
    invitano tutti i più grandi matematici per una conferenza di due mesi per creare l'intelligenza artificiale ma creano solo un programma che data la tesi crea la dimostrazione di un teorema (dimostrano leggi di Asimov) 
    1958 inventano una procedura di apprendimo automatico (nasce Lisp)
    1959 Arthur Samuel machine-learning (branca informatica che permette a una macchina di eseguire un compito senza essere programmata per farla (ingeniere-elettrotecnico)
##### anni 60 (Edward Albert Feigenbaum)
    nascono i sistemi esperti: software vengono scritte dal programmatore un insieme di regole per permettere alla macchina di fare ciò che deve (usate in medicina per fare diagnosi) oggi non esistono più
    era diffusa l'idea di macchine capaci di pensare o di macchina con coscenza

##### reti neurali
    rete fatta da tanti percettori (neuroni) si collegano a livelli:
      1) livelli input
      2) livelli interni/nascosti
      3) livelli output
    funziona che l'intelligenza/le conoscenze non stanno nei neuroni ma nelle connessioni (ogni volta che impariamo qualcosa si creano/distruggono delle connessioni quindi il nostro cervello viene modificato) = cervello umano
    chatgpt usa un'ordine di grandezza che supera il biolione di connessioni 
    si presuppone che in futuro si arriverà ad utilizzare una connessione che superi addirittura il cervello umano
